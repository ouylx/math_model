```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'D:\数学建模\集训4\cleaned_online_retail.csv')
```

```python
# 保留所需的列
data['CustomerID'] = data['CustomerID'].astype('int')
data1 = data[['CustomerID', 'StockCode']]
data1
```

```python
from collections import defaultdict

# 构建商品到客户的映射
item_customer_map = defaultdict(set)

for _, row in data1.iterrows():
    item_customer_map[row['StockCode']].add(row['CustomerID'])
```

```python
from itertools import combinations
import networkx as nx

# 创建无向图
G = nx.Graph()

# 遍历每个商品的客户列表，构建客户关系图
# 每条边的权重表示共同购买商品的次数
for customers in item_customer_map.values():
    for customer1, customer2 in combinations(customers, 2):
        if G.has_edge(customer1, customer2):
            G[customer1][customer2]['weight'] += 1
        else:
            G.add_edge(customer1, customer2, weight=1)

print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# 只保留度数大于某个阈值的节点
degrees = dict(G.degree())
subgraph_nodes = [node for node, degree in G.degree() if degree > 4000]
subgraph = G.subgraph(subgraph_nodes)
# 绘制网络图
plt.figure(figsize=(8, 8))
plt.rcParams['font.sans-serif'] = 'SimHei'  # 设置中文显示
plt.rcParams['axes.unicode_minus'] = False
# 使用spring_layout布局
pos = nx.spring_layout(subgraph, k=0.5)

# 计算中心性作为节点大小的权重
degree_centrality = nx.degree_centrality(subgraph)

# 使用中心性调整节点大小
node_sizes = [1000 * degree_centrality[n] for n in subgraph.nodes()]

# 绘制边，边宽与权重成比例
edge_widths = [d['weight'] * 0.01 for (u, v, d) in subgraph.edges(data=True)]

# 绘制图形，突出显示中心性较高的节点
plt.figure(figsize=(12, 12))
nx.draw_networkx(subgraph, pos, node_size=node_sizes, node_color='skyblue', with_labels=True, alpha=0.7, edge_color='gray',width=edge_widths)
#plt.savefig('客户关系网络图.png', dpi=500)
plt.show()
```

```python

features = {
    'avg_degree_centrality': nx.degree_centrality(G),  # 度中心性
    'avg_betweenness_centrality': nx.betweenness_centrality(G),  # 介数中心性
    'clustering_coefficient': nx.average_clustering(G)  # 平均聚s类系数
}
features
```

```python
betweenness_centrality = nx.betweenness_centrality(G)
print(f'平均介数中心性：{sum(betweenness_centrality.values())/len(betweenness_centrality)}')
```

```python
print(f'平均度数中心性：{sum(nx.degree_centrality(G).values())/len(nx.degree_centrality(G))}')
```

```python
ave_clustering = nx.average_clustering(G)
print(f'平均聚类系数：{ave_clustering}')
```

```python
# 计算网络直径（需要是连通图）
if nx.is_connected(G):
    print(f"网络直径: {nx.diameter(G)}")
else:
    print("Network is not connected, cannot compute diameter.")
```

```python
# 计算网络的平均最短路径长度
if nx.is_connected(G):
    print(f"网络的平均最短路径长度: {nx.average_shortest_path_length(G)}")
else:
    # 如果网络不连通，则计算每个连通分量的平均最短路径
    largest_cc = max(nx.connected_components(G), key=len)
    subgraph = G.subgraph(largest_cc)
    print(f"每个连通分量的平均最短路径: {nx.average_shortest_path_length(subgraph)}")
```

```python
# 使用节点度（degree centrality）来衡量客户的重要性
degree_centrality = nx.degree_centrality(G)

# 找出最重要的前五个客户
top_5_customers = sorted(degree_centrality, key=degree_centrality.get, reverse=True)[:5]
print("Top 5 Customers:", top_5_customers)
```

```python
# 统计每个商品的共现客户数量（即购买该商品的客户数）
item_customer_count = {item: len(customers) for item, customers in item_customer_map.items()}

# 找出被最多客户购买的前五个商品
top_5_items_by_customers = sorted(item_customer_count, key=item_customer_count.get, reverse=True)[:5]
print("Top 5 Items by Customer Count:", top_5_items_by_customers)
```

```python
# 比较第三问与第一问中的商品
first_question_top_5 = ['23843','23166','84077','22197','85099B']  # 假设第一问已经找出了这五个商品
common_items = set(top_5_items_by_customers).intersection(first_question_top_5)
print("Common Items:", common_items)

# 找出差异化商品
different_items = set(top_5_items_by_customers).difference(first_question_top_5)
print("Different Items:", different_items)
```

# 个性化推荐

```python
import pandas as pd

# 计算每个客户购买热销商品的总量
hot_items = ['23843','23166','84077','22197','85099B','22423', '85123A', '47566', '84879', '22720']  # 购买数量前五和购买客户数前五的并集
customer_hot_item_purchase = data[data['StockCode'].isin(hot_items)].groupby('CustomerID')['Quantity'].sum()

# 选择购买热销商品最多的客户群体
top_customers = customer_hot_item_purchase.nlargest(100).index  # 选择前100个客户作为特定群体
top_customers
```

```python
import networkx as nx

# 构建客户与商品的二分图
B = nx.Graph()
B.add_nodes_from(data['CustomerID'].unique(), bipartite=0)
B.add_nodes_from(data['StockCode'].unique(), bipartite=1)
edges = [(row['CustomerID'], row['StockCode']) for _, row in data.iterrows()]
B.add_edges_from(edges)

# 使用PageRank进行推荐
page_rank_scores = nx.pagerank(B)

# 为特定客户推荐商品
def pagerank_recommendations_for_customers(customers, page_rank_scores, top_k=5):
    recommendations = {}
    for customer in customers:
        customer_items = [item for item in B.neighbors(customer) if item in page_rank_scores]
        sorted_items = sorted(customer_items, key=lambda x: page_rank_scores[x], reverse=True)[:top_k]
        recommendations[customer] = sorted_items
    return recommendations

pagerank_recommendations = pagerank_recommendations_for_customers(top_customers, page_rank_scores)
```

# 演化网络

```python
from itertools import combinations
import networkx as nx
# 创建按月的时间窗口
data['Month'] = data['InvoiceDate'].dt.to_period('M')

# 第2步：使用groupby方法优化构建客户关系网络
def build_network_fast(month_data):
    G = nx.Graph()  # 创建空的无向图
    
    # 使用groupby方法按产品ID分组，构建客户之间的连接
    for product, group in month_data.groupby('StockCode'):
        customers = group['CustomerID'].unique()
        if len(customers) > 1:  # 至少有两个客户购买该商品
            # 创建所有客户对的组合，并计数
            for i, cust1 in enumerate(customers):
                for cust2 in customers[i + 1:]:
                    if G.has_edge(cust1, cust2):
                        G[cust1][cust2]['weight'] += 1
                    else:
                        G.add_edge(cust1, cust2, weight=1)
    return G
```

```python
# 第3步：分析网络特征，使用并行计算
def analyze_network_features(G):
    features = {
        'degree_centrality': nx.degree_centrality(G),  # 度中心性
        'betweenness_centrality': nx.betweenness_centrality(G),  # 介数中心性
        'clustering_coefficient': nx.average_clustering(G)  # 平均聚类系数
    }
    return features
```

```python
# 定义并行处理函数
def process_month(month):
    month_data = data[data['Month'] == month]
    G = build_network_fast(month_data)
    return month, G, analyze_network_features(G)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
def plot_net(G,month, threshold)
    # 只保留度数大于某个阈值的节点
    degrees = dict(G.degree())
    avg_degrees = sum(degrees.values())/len(degrees)
    subgraph_nodes = [node for node, degree in G.degree() if degree > threshold]
    subgraph = G.subgraph(subgraph_nodes)
    # 绘制网络图
    plt.figure(figsize=(12, 12))
    plt.rcParams['font.sans-serif'] = 'SimHei'  # 设置中文显示
    plt.rcParams['axes.unicode_minus'] = False
    # 使用spring_layout布局
    pos = nx.spring_layout(subgraph, k=0.5)
    
    # 计算中心性作为节点大小的权重
    degree_centrality = nx.degree_centrality(subgraph)
    
    # 使用中心性调整节点大小
    node_sizes = [1000 * degree_centrality[n] for n in subgraph.nodes()]
    
    # 绘制边，边宽与权重成比例
    edge_widths = [d['weight'] * 0.01 for (u, v, d) in subgraph.edges(data=True)]
    
    # 绘制图形，突出显示中心性较高的节点
    plt.figure(figsize=(12, 12))
    nx.draw_networkx(subgraph, pos, node_size=node_sizes, node_color='skyblue', with_labels=True, alpha=0.7, edge_color='gray',width=edge_widths)
    plt.savefig(f'{month}月客户关系网络图.png', dpi=500)
    plt.show()
```
