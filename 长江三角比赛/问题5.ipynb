```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import scipy.stats as stats
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv(
    r'C:\Users\25492\Desktop\2024年第四届长三角高校数学建模竞赛赛题\2024长三角选题B：人工智能范式的物理化学家\data.csv')
data
```

# 1. 变量选择

```python
X = data.iloc[:,5:] # x1-x100
Y = data.iloc[:,4]# class
Y
```

```python
# 套索回归选择变量
from sklearn.linear_model import LassoCV
# 数据预处理：标准化特征
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
 
# 使用LassoCV自带正则化路径长度和路径中的alpha个数来自动建立alpha选择的范围
ls_ = LassoCV(eps=0.0001,n_alphas=300,cv=5).fit(X_scaled,Y)
# 可以查看哪些特征的系数被压缩到零，实现了特征选择
non_zero_coef_indices = np.abs(ls_.coef_) != 0
selected_features = np.arange(X.shape[1])[non_zero_coef_indices]
print("Selected Features (indices): ", selected_features)
 
# 输出系数
coef = ls_.coef_
print("Lasso Coefficients: ", coef)
```

```python
X_selected = X.iloc[:,selected_features]
X_selected
```

```python
# 分析选择的变量之间的独立性
S_x = np.zeros([X_selected.shape[1],X_selected.shape[1]])
for i in range(X_selected.shape[1]):
    for j in range(X_selected.shape[1]):
        S_x[i,j] = stats.spearmanr(X_selected.iloc[:,i],X_selected.iloc[:,j])[0]
S_x = pd.DataFrame(S_x,index=X_selected.columns, columns=X_selected.columns)
#S_x.to_csv('X变量之间相关系数矩阵_class.csv')
S_x
```

```python
X_selected = X_selected.drop(['x4'], axis=1)
X_selected
```
